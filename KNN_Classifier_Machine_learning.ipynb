```python
# first of all import library 
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
# import dataset, you will find the dataset in my Github account
# the link are given bellow
dataset=pd.read_csv("C://Users//yasir//Desktop//Python//Youtube//Knn_dataset.csv")
dataset.head()
```

```python
# lets define X and y
# always in python number is started from 0
# in our dataset "age =0, salary =1", "purchased=2"
X=dataset.iloc[:,[0,1]].values
y=dataset.iloc[:,2]
```

```python
X
```

```python
y
```

```python
# now lets split our dataset into Training and testing
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
print("Train input data", X_train.shape)
print("Test input data",X_test.shape)
print("Train input data", y_train.shape)
print("Test input data",y_test.shape)
```

```python
# for better interpretation and standarized our data need to used this code
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
# now our dataset is read for train model witk KNN classifier
from sklearn.neighbors import KNeighborsClassifier
knn_classifier=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
knn_classifier.fit(X_train,y_train)
knn_classifier
```

```python
# lets predict the Y
y_predict=knn_classifier.predict(X_test)
y_predict
```

```python
# let check the confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predict)
```

```python
cm
```

```python
# check the accuarcy 
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predict)
```

```python
# visualising the training set results
# this code you will find in my github, i will directly copy and paste here
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, knn_classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('yellow', 'blue'))(i), label = j)
plt.title('Classifier KNN (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
# now visualising the test set results, copy the same code and change to test
from matplotlib.colors import ListedColormap
X_set, y_set = X_test, y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, knn_classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('green', 'blue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('black', 'white'))(i), label = j)
plt.title('Classifier KNN (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
